This notebook is an attempt of applying various Numpy operations on the image which was provided array of image in the form of text file and applying all the modification uisng numpy documentation.

```python
import numpy as np
```

```python
# Text file data converted to integer data type
img = np.loadtxt("/content/array-2.txt")
print(img)
```

```python
Q1 = img.astype('float64')
Q1
```

```python
Q2 = Q1 * 10
Q2
```

```python
arr = np.array([32, 54, 67, 88, 4])
total = np.sum(arr)
Q3 = Q2 + total
Q3
```

```python
Q4 = Q3 + Q3
Q4
```

```python
Tmp = np.arange(1, 812161, 1, dtype = 'float64')
Q5 = Q4 - Tmp
Q5
```

```python
Q6 = Q5.copy()
Q6[3] = 46225
Q6
```

```python
Q7 = Q6.copy()
for i in range(1000,200001):
  Q7[i] = np.add(Q7[i], 255)

for j in range(400000, 600001):
  Q7[j] = np.add(Q7[j], 255)
```

```python
Q8 = Q7.copy()
Q8 = np.reshape(Q8, (-1, 2))
Q8
```

```python
Q9 = Q8.copy()
Q9[:, [1,0]] = Q9[:, [0,1]]
Q9
```

```python
Q10 = Q9.copy()
Q10 = np.flip(Q10)
Q10 = np.fliplr(Q10)
Q10
```

```python
Q11 = Q10.copy()
Q11[100,:] = [46225, 353]
Q11
```

```python
Q12 = Q11.copy()
Q12[100000,0] = 46656
Q12
```

```python
Q13 = Q12.copy()
Q13[:,1] = np.subtract(Q13[:,1], 125)
Q13
```

```python
Q14 = Q13.copy()
Q14[:,0] = np.sqrt(Q14[:,0])
Q14
```

```python
Q14.shape
```

```python
Q15 = Q14.copy()
for i in range(406080):
  for j in range(2):
    if Q15[i,j] == 999:
      Q15[i,j] = 117
    

Q15
```

```python
Q16 = Q15.copy()
Q16 = np.transpose(Q16)
Q16[:,0] = np.subtract(Q16[:,0], 125)
Q16 = np.transpose(Q16)
Q16
```

```python
Q17 = Q16.copy()
Q17 = np.reshape(Q17, (423, 640, 3)) #using Numpy reshape function to reshape the array as per given dimension
Q17
```

```python
Q18 = Q17.copy()
Q18 = Q18.astype('int') #typecasting the array;s data type
Q18.dtype
```
